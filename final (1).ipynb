```python
#printing the states which are below poverty in 2011
avg_pov = data['2011 -Poverty'].mean()
less = data[data['2011 -Poverty']<avg_pov]
print(less['States_Union Territories'])
```

```python
#printing state having maximum literacy rate
max_lit = data['2011- LIT'].max()
max_lit_row = data[data['2011- LIT'] == max_lit]
print(max_lit_row['States_Union Territories'])
```

```python
#printing state having minimum literacy rate
min_lit = data['2011- LIT'].min()
min_lit_row = data[data['2011- LIT'] == min_lit]
print(min_lit_row['States_Union Territories'])
```

```python
unemp2001 = '2001 -UNEMP'
unemp2011 = '2011 -UNEMP'
diff = data[unemp2011] - data[unemp2001]
rel = pd.concat([data['States_Union Territories'],diff])
print(rel)
```

```python
#printing states and sex ratio below the average sex ratio in 2011
avg_sex_ratio = data['2011 -SEX_Ratio'].mean()
less = data[data['2011 -SEX_Ratio']<avg_sex_ratio]
print("Average sex ratio in 2011 is",avg_sex_ratio)
print(less[['States_Union Territories','2011 -SEX_Ratio']])
```

**PLOTING FUNCTIONS**

```python
data.plot()
```

```python
#BAR GRAPH
pop_2001 = data['2001 - POP']
pop_2011 = data['2011- POP']
plt.figure(figsize=(10, 6))
plt.bar('2001', pop_2001, color='blue', alpha=0.5, label='2001')
plt.bar('2011', pop_2011, color='red', alpha=0.5, label='2011')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Comparison of Population in 2001 and 2011')
plt.legend()
plt.show()
```

```python
#SCATTER PLOT
data.plot(kind = 'scatter', x = '2011- POP', y = '2011- LIT')
plt.show()
```

```python
#HISTOGRAPH
pop_2001 = data['2001 - LIT']
pop_2011 = data['2011- LIT']
plt.figure(figsize=(10, 6))
plt.hist(pop_2001, bins=10,  alpha=0.5, label='2001')
plt.hist(pop_2011, bins=10,  alpha=0.5, label='2011')
plt.xlabel('literacy')
plt.title('Literacy in 2001 and 2011')
plt.legend()
plt.show()
```

```python
#correlation
df_corr_11 = data[['2011-12-INC','2011- LIT','2011- POP','2011 -SEX_Ratio','2011 -UNEMP','2011 -Poverty']].corr()

#plotting in seaborn
plt.figure(figsize=(10,7))
sns.heatmap(df_corr_11,cmap='Blues',annot=True,linewidths=0.2,linecolor='white')
plt.title('correlation')
plt.tight_layout()
```
